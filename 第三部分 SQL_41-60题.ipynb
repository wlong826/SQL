### 48. å°†æ‰€æœ‰è·å–å¥–é‡‘çš„å‘˜å·¥å½“å‰çš„è–ªæ°´å¢åŠ 10%

ã€€ã€€è¯·ä½ å†™å‡ºæ›´æ–°è¯­å¥ï¼Œå°†æ‰€æœ‰è·å–å¥–é‡‘çš„å‘˜å·¥å½“å‰çš„(salaries.to_date='9999-01-01')è–ªæ°´å¢åŠ 10%ã€‚(emp_bonusé‡Œé¢çš„emp_noéƒ½æ˜¯å½“å‰è·å¥–çš„æ‰€æœ‰å‘˜å·¥)ï¼›

```python
# æ–¹æ³•ä¸€ï¼šå­æŸ¥è¯¢
update salaries s
set s.salary = s.salary * 1.1
where s.to_date = '9999-01-01'
and s.emp_no in (
					select emp_no
					from emp_bonus
				);
```

```python
# æ–¹æ³•äºŒï¼šè¿æ¥æŸ¥è¯¢
update salaries as s 
join emp_bonus as e
on s.emp_no = e.emp_no
set salary = salary*1.1
where to_date = '9999-01-01';
```

### 54. æŸ¥æ‰¾æ’é™¤æœ€å¤§ã€æœ€å°salaryä¹‹åçš„å½“å‰å‘˜å·¥çš„å¹³å‡å·¥èµ„avg_salary

### 59. è·å–æœ‰å¥–é‡‘çš„å‘˜å·¥ç›¸å…³ä¿¡æ¯(â˜…â˜…â˜…â˜…)

ã€€ã€€ç»™å‡ºemp_noã€first_nameã€last_nameã€å¥–é‡‘ç±»å‹btypeã€å¯¹åº”çš„å½“å‰è–ªæ°´æƒ…å†µsalaryä»¥åŠå¥–é‡‘é‡‘é¢bonusã€‚ bonusç±»å‹btypeä¸º1å…¶å¥–é‡‘ä¸ºè–ªæ°´salaryçš„10%ï¼Œbtypeä¸º2å…¶å¥–é‡‘ä¸ºè–ªæ°´çš„20%ï¼Œå…¶ä»–ç±»å‹å‡ä¸ºè–ªæ°´çš„30%ã€‚ å½“å‰è–ªæ°´è¡¨ç¤ºto_date='9999-01-01'ï¼›

```python
select
e.emp_no
,first_name
,last_name
,btype
,salary
,case btype when 1 then 0.1*salary
when 2 then 0.2*salary 
else 0.3*salary end bonus
from employees e
join emp_bonus e_b
on e.emp_no = e_b.emp_no
join salaries s
on e.emp_no = s.emp_no
where s.to_date = '9999-01-01';
```

### 60. ç»Ÿè®¡salaryçš„ç´¯è®¡å’Œrunning_total(â˜…â˜…â˜…â˜…)

ã€€ã€€æŒ‰ç…§salaryçš„ç´¯è®¡å’Œrunning_totalï¼Œå…¶ä¸­running_totalä¸ºå‰Nä¸ªå½“å‰( to_date = '9999-01-01')å‘˜å·¥çš„salaryç´¯è®¡å’Œï¼Œå…¶ä»–ä»¥æ­¤ç±»æ¨ã€‚ å…·ä½“ç»“æœå¦‚ä¸‹Demoå±•ç¤ºï¼›

```python
select
emp_no
,salary
,sum(salary) over(order by emp_no) running_total
from salaries
where to_date = '9999-01-01';
```
