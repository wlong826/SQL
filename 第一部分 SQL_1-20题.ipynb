### 3. 	æŸ¥æ‰¾å½“å‰è–ªæ°´è¯¦æƒ…ä»¥åŠéƒ¨é—¨ç¼–å·dept_no

### 11. è·å–æ‰€æœ‰å‘˜å·¥å½“å‰çš„manager

### 12. è·å–æ¯ä¸ªéƒ¨é—¨ä¸­å½“å‰å‘˜å·¥è–ªæ°´æœ€é«˜çš„ç›¸å…³ä¿¡æ¯(â˜…â˜…â˜…â˜…â˜…)

ã€€ã€€ç»™å‡ºdept_no, emp_noä»¥åŠå…¶å¯¹åº”çš„salaryï¼ŒæŒ‰ç…§éƒ¨é—¨ç¼–å·å‡åºæ’åˆ—

```python
#æ€è·¯1ï¼šçª—å£å‡½æ•°
select
dept_no
,emp_no
,salary
from 
(
	select
	dept_no
	,d.emp_no
    ,salary
	,row_number() over(partition by dept_no order by salary desc) s_rank
	from dept_emp d
	join salaries s
	on d.emp_no = s.emp_no
) t
where t.s_rank = 1
```

```python
# æ€è·¯2ï¼šé‡‡ç”¨group byåˆ†ç»„ + å†…è¿æ¥
select
z.dept_no
,d.emp_no
,z.sa
from dept_emp d
join salaries s
on d.emp_no = s.emp_no
join 
(
	select
	dept_no
	,max(salary) sa
	from dept_emp d
	join salaries s
	on d.emp_no = s.emp_no
	group by dept_no
) z
on d.dept_no = z.dept_no
and s.salary = z.sa
order by z.dept_no
```

### 16. ç»Ÿè®¡å‡ºå„ä¸ªtitleç±»å‹å¯¹åº”çš„å‘˜å·¥è–ªæ°´å¯¹åº”çš„å¹³å‡å·¥èµ„avg

ã€€ã€€ç»Ÿè®¡å‡ºå„ä¸ªtitleç±»å‹å¯¹åº”çš„å‘˜å·¥è–ªæ°´å¯¹åº”çš„å¹³å‡å·¥èµ„avgï¼Œç»“æœç»™å‡ºtitleä»¥åŠå¹³å‡å·¥èµ„avgï¼Œå¹¶ä¸”ä»¥avgå‡åºæ’åºï¼›

```python
select
title
,avg(salary) avg
from titles t
join salaries s
on t.emp_no = s.emp_no
group by title
order by avg
```

### 17. è·å–è–ªæ°´ç¬¬äºŒå¤šçš„å‘˜å·¥çš„emp_noä»¥åŠå…¶å¯¹åº”çš„è–ªæ°´salary

### 18. æŸ¥æ‰¾è–ªæ°´æ’åç¬¬äºŒå¤šçš„å‘˜å·¥ç¼–å·emp_noã€è–ªæ°´salaryã€last_nameä»¥åŠfirst_nameï¼Œä¸èƒ½ä½¿ç”¨order byå®Œæˆ(â˜…â˜…â˜…â˜…)

```python
# æ–¹æ³•ä¸€ï¼šå…ˆæŸ¥æ‰¾æœ€é«˜è–ªæ°´ï¼Œç„¶ååœ¨æŸ¥å‡ºç¬¬äºŒé«˜å·¥èµ„ï¼Œæœ€åå°†å°†ç¬¬äºŒé«˜å·¥èµ„ä½œä¸ºæŸ¥è¯¢æ¡ä»¶
select s.emp_no, s.salary, e.last_name, e.first_name
from salaries s join employees e
on s.emp_no = e.emp_no
where s.salary = (                            # æœ€åï¼Œå°†ç¬¬äºŒé«˜å·¥èµ„ä½œä¸ºæŸ¥è¯¢æ¡ä»¶
			    	select max(salary)        # å…¶æ¬¡ï¼ŒæŸ¥å‡ºé™¤æœ€é«˜è–ªæ°´ä¹‹å¤–çš„æœ€é«˜å·¥èµ„(ç¬¬äºŒé«˜å·¥èµ„)
				    from salaries
				    where salary < (
								        select max(salary)    # é¦–å…ˆï¼ŒæŸ¥å‡ºæœ€é«˜è–ªæ°´
								        from salaries
							        )
			    )
```

```python
# æ–¹æ³•äºŒï¼š
select *
from
(
    select 
    e.emp_no
    ,salary
    ,last_name
    ,first_name
    from employees e
    join salaries s
    on e.emp_no = s.emp_no
    where e.emp_no not in (
					        select emp_no
					        from salaries
					        where salary >= ALL(select salary from salaries)
						  )
) t
where salary >= ALL(
                        select salary
					    from salaries s
					    where emp_no not in (
										        select emp_no
										        from salaries
										        where salary >= ALL(select salary from salaries)
											  )
					)
```

```python
# æ–¹æ³•ä¸‰ï¼šå‚è€ƒç­”æ¡ˆé‡Œï¼Œè‡ªè¿æ¥
select s.emp_no, s.salary, e.last_name, e.first_name
from salaries s join employees e
on s.emp_no = e.emp_no
where s.salary = (
				    select s1.salary
				    from salaries s1 join salaries s2      -- è‡ªè¿æ¥æŸ¥è¯¢
				    on s1.salary <= s2.salary
				    group by s1.salary                     -- å½“s1<=s2é“¾æ¥å¹¶ä»¥s1.salaryåˆ†ç»„æ—¶ä¸€ä¸ªs1ä¼šå¯¹åº”å¤šä¸ªs2
				    having count(distinct s2.salary) = 2   -- (å»é‡ä¹‹åçš„æ•°é‡å°±æ˜¯å¯¹åº”çš„åæ¬¡)
				  )
```
