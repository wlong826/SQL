### 3. 	查找当前薪水详情以及部门编号dept_no

　　查找各个部门当前领导的薪水详情以及其对应部门编号dept_no,输出结果以salaries.emp_no升序排序，并且请注意输出结果里面dept_no列是最后一列；

```python
select 
s.emp_no
,s.salary
,s.from_date
,s.to_date
,m.dept_no
from dept_manager m
join salaries s
on m.emp_no = s.emp_no
order by s.emp_no
```

### 11. 获取所有员工当前的manager

　　获取所有的员工和员工对应的经理，如果员工本身是经理的话则不显示；

```python
select 
e.emp_no
,d.emp_no manager
from dept_emp e
join dept_manager d
on e.dept_no = d.dept_no
where e.emp_no <> d.emp_no # <> 即 !=
```

### 12. 获取每个部门中当前员工薪水最高的相关信息(★★★★★)

　　给出dept_no, emp_no以及其对应的salary，按照部门编号升序排列

```python
#思路1：窗口函数
select
dept_no
,emp_no
,salary
from 
(
	select
	dept_no
	,d.emp_no
    ,salary
	,row_number() over(partition by dept_no order by salary desc) s_rank
	from dept_emp d
	join salaries s
	on d.emp_no = s.emp_no
) t
where t.s_rank = 1
```

```python
# 思路2：采用group by分组 + 内连接
select
z.dept_no
,d.emp_no
,z.sa
from dept_emp d
join salaries s
on d.emp_no = s.emp_no
join 
(
	select
	dept_no
	,max(salary) sa
	from dept_emp d
	join salaries s
	on d.emp_no = s.emp_no
	group by dept_no
) z
on d.dept_no = z.dept_no
and s.salary = z.sa
order by z.dept_no
```

### 15. 查找employees表emp_no与last_name的员工信息

### 16. 统计出各个title类型对应的员工薪水对应的平均工资avg

　　统计出各个title类型对应的员工薪水对应的平均工资avg，结果给出title以及平均工资avg，并且以avg升序排序；

```python
select
title
,avg(salary) avg
from titles t
join salaries s
on t.emp_no = s.emp_no
group by title
order by avg
```

### 17. 获取薪水第二多的员工的emp_no以及其对应的薪水salary

### 18. 查找薪水排名第二多的员工编号emp_no、薪水salary、last_name以及first_name，不能使用order by完成(★★★★)

```python
# 方法一：先查找最高薪水，然后在查出第二高工资，最后将将第二高工资作为查询条件
select s.emp_no, s.salary, e.last_name, e.first_name
from salaries s join employees e
on s.emp_no = e.emp_no
where s.salary = (                            # 最后，将第二高工资作为查询条件
			    	select max(salary)        # 其次，查出除最高薪水之外的最高工资(第二高工资)
				    from salaries
				    where salary < (
								        select max(salary)    # 首先，查出最高薪水
								        from salaries
							        )
			    )
```

```python
# 方法二：
select *
from
(
    select 
    e.emp_no
    ,salary
    ,last_name
    ,first_name
    from employees e
    join salaries s
    on e.emp_no = s.emp_no
    where e.emp_no not in (
					        select emp_no
					        from salaries
					        where salary >= ALL(select salary from salaries)
						  )
) t
where salary >= ALL(
                        select salary
					    from salaries s
					    where emp_no not in (
										        select emp_no
										        from salaries
										        where salary >= ALL(select salary from salaries)
											  )
					)
```

```python
# 方法三：参考答案里，自连接
select s.emp_no, s.salary, e.last_name, e.first_name
from salaries s join employees e
on s.emp_no = e.emp_no
where s.salary = (
				    select s1.salary
				    from salaries s1 join salaries s2      -- 自连接查询
				    on s1.salary <= s2.salary
				    group by s1.salary                     -- 当s1<=s2链接并以s1.salary分组时一个s1会对应多个s2
				    having count(distinct s2.salary) = 2   -- (去重之后的数量就是对应的名次)
				  )
```
